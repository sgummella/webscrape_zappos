```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
from wordcloud import WordCloud
plt.style.use('ggplot')
```

```python
zappos = pd.read_csv('zappos_reviews.csv')
zappos.groupby(['Brand','Product'])
zappos.describe()
```

```python
zappos = zappos.fillna("")
zappos['Review_text'] = [elem.strip('\r\n') for elem in zappos['Review_text']]


```

```python
zappos = zappos.sort_values(by = ['True_size_feeling','True_width_feeling','Arch_support'], ascending=False)
zappos

```

```python
zappos.loc[zappos.Product == 'Chuck Taylor® All Star® Core Ox','Arch_support'] = 52
zappos.loc[zappos.Product == 'Addiction™ Walker', 'Arch_support'] = 53
zappos.loc[zappos.Product == 'Chuck Taylor® All Star® Core Hi', 'Arch_support'] = 75
zappos.loc[zappos.Product == 'Chuck Taylor All Star Seasonal Ox', 'Arch_support'] = 63
zappos.loc[zappos.Product == 'Chuck Taylor® All Star® Leather Ox', 'Arch_support'] = 52
```

```python
brand_price = zappos[['Brand', 'Price']]
bp_mean = brand_price.groupby(brand_price['Brand']).mean()
bp_mean = bp_mean.reset_index()

bar_bp_mean = plt.barh(bp_mean['Brand'], bp_mean['Price'])
plt.title("Mean Price by Brand")
plt.xlabel('Price')
plt.ylabel('Brand')
plt.xticks(np.arange(0,180,20))
```

```python
from wordcloud import WordCloud
Vans = zappos[zappos['Brand'] == 'Vans']
Vanswc = WordCloud(background_color="white", max_words=2000, width=800, height=400)
# generate word cloud
Vanswc.generate(' '.join(Vans['Review_text']))
plt.figure(figsize=(12, 6))
plt.imshow(Vanswc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
